```python
path = input("/content/compressed hpe doc.pdf")#"C:/Users/Sourav/Downloads/Naïve Bayes.pdf"
loader = PyPDFLoader(path)
pages = loader.load()
```

```python
#number of pages
len(pages)
```

```python
pages[0]
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
docs = splitter.split_documents(pages)
```

```python
len(docs)
```

```python
embeddings = HuggingFaceEmbeddings()
doc_search = Chroma.from_documents(docs, embeddings)
```

```python
query = "What is HPE"
similar_docs = doc_search.similarity_search(query, k=3)
```

```python
similar_docs
```

```python
repo_id = "tiiuae/falcon-7b"
llm = HuggingFaceHub(huggingfacehub_api_token = os.environ['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'],
                     repo_id=repo_id, model_kwargs={'temperature': 0.2, 'max_length':1000})

#you can modify the parameters temperature and max length to generate different responses

# from langchain_community.llms.huggingface_endpoint import HuggingFaceEndpoint

# repo_id = "tiiuae/falcon-7b"
# llm = HuggingFaceEndpoint(huggingfacehub_api_token=os.environ['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'],
#                           repo_id=repo_id, model_kwargs={'temperature': 0.2, 'max_length':1000})
```

```python


retrieval_chain = RetrievalQA.from_chain_type(
    llm,
    chain_type='stuff',
    retriever=doc_search.as_retriever()
)
```
